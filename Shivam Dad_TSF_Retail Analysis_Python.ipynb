# Describing and Explaining Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats  as ss
from scipy.stats import trim_mean
```

```python
data = pd.read_csv(r"C:\Users\Nalini Daad\Desktop\Internships\Summer Internships\GRIP The Sparks Foundation\eda retail\SampleSuperstore.csv")
print("Top\n\n", data.head(10))
print("\n\nBottom\n\n", data.tail(10))
```

```python
# removing the duplicate values

newdata = data.drop_duplicates()
```

```python
newdata.shape
```

```python
newdata.info()
```

```python
newdata.describe()
```

```python
# After discarding top and bottom 20 % values eliminating outliers
print("Sales mean : ", newdata.Sales.mean())

Sales_TM = trim_mean(newdata.Sales, 0.2)
print ("\nSales trimmed mean: ", Sales_TM)

print("\nProfit mean : ", newdata.Profit.mean())

Profit_TM = trim_mean(newdata.Profit, 0.2)
print ("\nProfit trimmed mean: ", Profit_TM)

#standard deviation


print("\nVariance : " , newdata.Sales.var())
print("\nStandard deviation : " , newdata.Sales.std())
```

```python
newdata.corr() # value more than 0 shows that the it has +ve relation between 2 varivbles, else if -ve then both the variables go in opposite direction
```

```python
newdata.cov() # +ve -> means trendline moving in upward direction, -ve -> means trendline going in downward direction 
```

# Data Visualization

```python
# Histogram 
newdata.hist(bins=50, figsize=(15,15))
```

```python
# Boxplot
x=list(newdata.Sales)  # large number of outliers on positive side
y=list(newdata.Profit)   # large number of outliers on both positive and negative side
z=list(newdata.Quantity) # qutliers above 9
a=list(newdata.Discount)  # outliers above 5, maximum discount between 0.0-0.2
plt.boxplot(x)
plt.show()
plt.boxplot(y)
plt.show()
plt.boxplot(z)
plt.show()
plt.boxplot(a)
plt.show()
```

```python
# Scatter plot
plt.scatter(x,y, label= "stars",   # mostly our profits increase with the Sales
            marker= "*", s=30)     # at some Slaes profit dec. due to higher Discounts   
plt.xlabel('Sales')
# frequency label
plt.ylabel('Profit')
# function to show the plot
plt.show()
```

```python
# Bar charts

newdata.plot.bar()
  
# plot between 2 attributes
plt.bar(newdata['Sub-Category'], newdata['Category'])
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
# countplot
print(newdata['Segment'].value_counts())
plt.figure(figsize= (10,5))
sns.countplot(x=newdata['Segment'])
plt.show()
```

```python
# Pair plot
figsize=(20,15)
sns.pairplot(newdata, hue = 'Segment')
plt.show()
```
